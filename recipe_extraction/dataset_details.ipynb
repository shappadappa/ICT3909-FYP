# **Dataset Details**

This notebook provides a summary of some of the details of the dataset.

This notebook requires downloading the `structured_recipes.json` file, and running it through `data_supplementation.ipynb`, or simply downloading `supplemented_structured_recipes.json` to get the ready-supplemented dataset.

```python
import json
from collections import Counter
```

## Recipes Details

```python
with open("./supplemented_structured_recipes.json", "r") as file:
    recipes = json.load(file)
```

```python
print(f"Number of recipes: {len(recipes)}")
```

```python
dietary_restrictions_recipes_counter = Counter()
most_common_ingredients = Counter()

for recipe in recipes:
    for dietary_tag in recipe["dietary_tags"]:
        dietary_restrictions_recipes_counter[dietary_tag] += 1

    for ingredient_name_and_quantity in recipe["ingredients"]:
        most_common_ingredients[ingredient_name_and_quantity["ingredient"]] += 1
```

```python
print("Number of recipes by dietary restriction:")
for key, value in dietary_restrictions_recipes_counter.most_common():
    print(f"\t- {key}: {value}")
```

```python
print("20 most commonly used ingredients in recipes, by frequency:")
for key, value in most_common_ingredients.most_common(20):
    print(f"\t- {key}: {value}")
```

## Ingredients Details

```python
with open("./supplemented_structured_ingredients.json", "r") as file:
    ingredients = json.load(file)
```

```python
print(f"Number of ingredients: {len(ingredients)}")
```

```python
null_ni_attribute_ingredients_counter = Counter()
nutritional_information_keys = ["calories", "carbohydrates", "sugar", "protein", "fat", "saturated_fat", "fiber", "sodium"]

dietary_restrictions_ingredients_counter = Counter()

for ingredient_name_and_quantity in ingredients:
    for key in nutritional_information_keys:
        if ingredient_name_and_quantity["nutritional_information"] [key] is None:
            null_ni_attribute_ingredients_counter[key] += 1

    if ingredient_name_and_quantity ["nutritional_information"]["is_gluten_free"] is True:
        dietary_restrictions_ingredients_counter["gluten_free"] += 1
    if ingredient_name_and_quantity ["nutritional_information"]["is_lactose_free"] is True:
        dietary_restrictions_ingredients_counter["lactose_free"] += 1
    if ingredient_name_and_quantity ["nutritional_information"]["is_vegetarian"] is True:
        dietary_restrictions_ingredients_counter["vegetarian"] += 1
    if ingredient_name_and_quantity ["nutritional_information"]["is_vegan"] is True:
        dietary_restrictions_ingredients_counter["vegan"] += 1
```

```python
print("Number of ingredients with null nutritional information attributes:")
for key, count in null_ni_attribute_ingredients_counter.most_common():
    print(f"\t- {key}: {count}")
```

```python
print("Number of ingredients by dietary restriction:")
for key, value in dietary_restrictions_ingredients_counter.most_common():
    print(f"\t- {key}: {value}")
```
